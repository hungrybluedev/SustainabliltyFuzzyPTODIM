```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
plt.rcParams["figure.dpi"] = 1000
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = '8'
```

```python
ranking_data = pd.read_csv('weight_rankings.csv')
ranking_data
```

```python
x = ranking_data.columns

plt.xticks(rotation='vertical')
plt.yticks(ticks=range(1, len(x) + 1))

plt.xlabel('Alternatives')
plt.ylabel('Rank')

for _, row in ranking_data.iterrows():
    plt.scatter(x=x, y=list(row), alpha=0.002, c='black')
```
